**code des points seulement**

```python
import geopandas as gpd
from shapely.geometry import Point
import random
import pyarrow as pa
import pyarrow.parquet as pq
import pandas as pd

# Charger le polygone  partir du fichier shapefile
path_to_shapefile = "MAROC.shp"
polygons = gpd.read_file(path_to_shapefile)

# Assurez-vous qu'il y a un seul polygone
if len(polygons) != 1:
    raise ValueError("Le fichier shapefile ne contient pas un seul polygone.")

# Slectionner le polygone unique
polygon = polygons.geometry.iloc[0]

# Dfinir les limites du polygone pour la gnration alatoire
min_x, min_y, max_x, max_y = polygon.bounds

# Gnrer 1000 points alatoires  l'intrieur du polygone
random_points = []
while len(random_points) < 1000:
    # Gnrer des coordonnes x et y alatoires  l'intrieur des limites du polygone
    random_point = Point(random.uniform(min_x, max_x), random.uniform(min_y, max_y))
    # Vrifier si le point est  l'intrieur du polygone
    if polygon.contains(random_point):
        random_points.append(random_point.wkt)  # Convertir la gomtrie en format WKT

# Crer un DataFrame Pandas  partir des points alatoires
df = pd.DataFrame({'geometry': random_points})

# Crer un schma PyArrow pour la colonne de gomtries
schema = pa.schema([('geometry', pa.string())])

# Crer une table PyArrow  partir du DataFrame
table = pa.Table.from_pandas(df, schema=schema)

# crire la table PyArrow dans un fichier GeoParquet
path_to_output_geoparquet = "OUTPUT1000.geoparquet"
pq.write_table(table, path_to_output_geoparquet)
```

```python
import geopandas as gpd
from shapely.geometry import Point
import random
import pyarrow as pa
import pyarrow.parquet as pq
import pandas as pd
import datetime
import string

# Charger le polygone  partir du fichier shapefile
path_to_shapefile = "MAROC.shp"
polygons = gpd.read_file(path_to_shapefile)

# Assurez-vous qu'il y a un seul polygone
if len(polygons) != 1:
    raise ValueError("Le fichier shapefile ne contient pas un seul polygone.")
polygon = polygons.geometry.iloc[0]

# Obtenir le CRS du polygone
crs = polygons.crs

# Dfinir les limites du polygone pour la gnration alatoire
min_x, min_y, max_x, max_y = polygon.bounds

# Gnrer 1000 points alatoires  l'intrieur du polygone
random_points = []
while len(random_points) < 1000:
    random_point = Point(random.uniform(min_x, max_x), random.uniform(min_y, max_y))
    if polygon.contains(random_point):
        random_points.append(random_point)

# Dfinir les colonnes d'attributs pour chaque point
attrib_data = {
    'propriete4': [datetime.datetime(2022, 1, 1) + datetime.timedelta(days=random.randint(1, 365)) for _ in range(1000)],
    'propriete1': [''.join(random.choice(string.ascii_letters) for _ in range(10))for _ in range(1000)],
    'propriete2': [random.uniform(-10, 10) for _ in range(1000)],
    'propriete3': [random.uniform(-10, 10) for _ in range(1000)],
    'attribut1jour0': [random.uniform(0, 100) for _ in range(1000)],
    'attribut1jour-1': [random.uniform(0, 100) for _ in range(1000)],
    'attribut1jour-2': [random.uniform(0, 100) for _ in range(1000)],
    'attribut1jour-3': [random.uniform(0, 100) for _ in range(1000)],
    'attribut1jour-4': [random.uniform(0, 100) for _ in range(1000)],
    'attribut1jour-5': [random.uniform(0, 100) for _ in range(1000)],
    'attribut1jour-6': [random.uniform(0, 100) for _ in range(1000)],
    'attribut2jour0': [random.uniform(0, 20) for _ in range(1000)],
    'attribut2jour-1': [random.uniform(0, 20) for _ in range(1000)],
    'attribut2jour-2': [random.uniform(0, 20) for _ in range(1000)],
    'attribut2jour-3': [random.uniform(0, 20) for _ in range(1000)],
    'attribut2jour-4': [random.uniform(0, 20) for _ in range(1000)],
    'attribut2jour-5': [random.uniform(0, 20) for _ in range(1000)],
    'attribut2jour-6': [random.uniform(0, 20) for _ in range(1000)],
    'attribut3jour0': [random.uniform(0, 50) for _ in range(1000)],
    'attribut3jour-1': [random.uniform(0, 50) for _ in range(1000)],
    'attribut3jour-2': [random.uniform(0, 50) for _ in range(1000)],
    'attribut3jour-3': [random.uniform(0, 50) for _ in range(1000)],
    'attribut3jour-4': [random.uniform(0, 50) for _ in range(1000)],
    'attribut3jour-5': [random.uniform(0, 50) for _ in range(1000)],
    'attribut3jour-6': [random.uniform(0, 50) for _ in range(1000)],
}

# Crer un DataFrame Pandas avec les donnes d'attributs
attrib_df = pd.DataFrame(attrib_data)

# Convertir les colonnes de texte en Unicode
string_columns = attrib_df.select_dtypes(include=['object']).columns
attrib_df[string_columns] = attrib_df[string_columns].astype('U')

attrib_data['geometry'] = random_points
attrib_df = pd.DataFrame(attrib_data)

# Convertir en GeoDataFrame
gdf = gpd.GeoDataFrame(attrib_df, geometry='geometry', crs=crs)

# crire le GeoDataFrame dans un fichier GeoParquet
path_to_output_geoparquet = "OUTPUT1000.geoparquet"
gdf.to_parquet(path_to_output_geoparquet, engine='pyarrow')

```

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
import re

# Charger les donnes gnres alatoirement depuis le fichier GeoParquet
path_to_output_geoparquet = "OUTPUT1000.geoparquet"
data = pd.read_parquet(path_to_output_geoparquet)

# Convertir les donnes de gomtrie en chanes de caractres
data['geometry'] = data['geometry'].apply(lambda x: x.decode('latin-1'))

# Fonction pour extraire les coordonnes et crer les objets Point
def create_point_geometry(row):
    match = re.search(r'POINT \((-?\d+\.\d+) (-?\d+\.\d+)\)', row)
    if match:
        lon, lat = map(float, match.groups())  # Extraire les coordonnes
        return Point(lon, lat)
    else:
        return None

# Convertir les coordonnes en objets Point pour la colonne 'geometry'
data['geometry'] = data['geometry'].apply(create_point_geometry)

# Convertir les colonnes datetime en chanes de caractres
datetime_columns = data.select_dtypes(include=['datetime64']).columns
data[datetime_columns] = data[datetime_columns].astype(str)

# Crer un GeoDataFrame avec la colonne 'geometry'
gdf = gpd.GeoDataFrame(data, geometry='geometry')

# Enregistrer le GeoDataFrame en tant que fichier shapefile
path_to_output_shapefile = "OUTPUT1000.shp"
gdf.to_file(path_to_output_shapefile)
```

```python
import streamlit as st
import geopandas as gpd
import streamlit.components.v1 as components

# Titre de l'application Streamlit
st.title("Visualisation des donnes GeoParquet")

# Charger les donnes GeoParquet
path_to_output_geoparquet = "OUTPUT1000.geoparquet"  # Mettez ici le chemin absolu vers votre fichier
data = gpd.read_parquet(path_to_output_geoparquet)

# Sidebar pour la slection de la colonne
selected_column = st.sidebar.selectbox("Slectionner la colonne  cartographier", data.columns)

# Afficher les options pour la cartographie thmatique
theme_mapping_options = st.sidebar.checkbox("Activer la cartographie thmatique", value=True)
if theme_mapping_options:
    theme_column = st.sidebar.selectbox("Choisir la colonne pour la cartographie thmatique", data.columns)

# Afficher la carte
st_geospatial_figure = components.declare_component(
    "st_geospatial_figure",
    path="st_geospatial_figure",
)

# Convertir la colonne slectionne en GeoJSON
geojson_data = data[[selected_column, 'geometry']].to_json()

# Configurer les options pour la carte gospatiale
config = {
    "theme_mapping": theme_mapping_options,
    "theme_column": theme_column if theme_mapping_options else None,
}

# Afficher la carte gospatiale
result = st_geospatial_figure(geojson_data, config=config, key="geojson_data")

# Afficher les informations lies  la carte (par exemple, popup au survol)
if result is not None:
    st.json(result)
```
